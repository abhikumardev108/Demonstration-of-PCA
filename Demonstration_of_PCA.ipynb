```python
import numpy as np
import pandas as pd
```

```python
data = np.array([[3,7],
                 [-4,-6],
                 [1,-1],
                 [7,8],
                 [-4,-1],
                 [-3,-7]])
```

```python
dataframe = pd.DataFrame(data, columns=['x1', 'x2'])
```

Step 1. --> Standard Normal form : mean = 0 && standard deviation = 1

```python
dataframe
```

```python
dataframe.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataScaled = scaler.fit_transform(dataframe)
```

```python
type(dataScaled)
```

```python
dataframe = pd.DataFrame(dataScaled, columns=['x1', 'x2'])
```

```python
dataframe.describe()
```

Step 2. --> Covariance matrix between the two features that we have (x1,x2)

```python
## Approach 1.
c1 = dataframe.x1
c2 = dataframe.x2
np.cov(c1,c2)
```

```python
## Approach 2.
covarianceMatrix = dataframe.T @ dataframe/5
covarianceMatrix
```

Step 3. --> Evaluate the eigan value and eigan vector of the above covariance matrix.

```python
eigenValues, eigenVectors = np.linalg.eig(covarianceMatrix)
```

```python
## eigenValues = features
## eigenValues represents the strength of the info. given by the eigenVector


## Highest eigenValues --> takes for creating PCA1 and so on.
eigenValues
```

```python
eigenVectors
```

```python
eigenVectors[:, 0]
```

```python
dataframe
```

```python
## PCA1 --> contains the maximum info. of the original two features that u have.

PCA1 = dataframe @ eigenVectors[:, 0]
PCA1
```

```python
PCA2 = dataframe @ eigenVectors[:, 1]   ## @ -> multiply
PCA2
```
